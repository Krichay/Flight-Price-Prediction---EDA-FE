```python
### Feature Engineering

df_final['date'] = df_final['Date_of_Journey'].str.split('/').str[0]
df_final['month'] = df_final['Date_of_Journey'].str.split('/').str[1]
df_final['year'] = df_final['Date_of_Journey'].str.split('/').str[2]

#df_final['date'] = df_final['Date_of_Journey'].apply(lambda x:x.split('/')[0])
#df_final['month'] = df_final['Date_of_Journey'].apply(lambda x:x.split('/')[1])
#df_final['year'] = df_final['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df_final.head()
```

```python
df_final['date'] = df_final['date'].astype(int)
df_final['month'] = df_final['month'].astype(int)
df_final['year'] = df_final['year'].astype(int)
```

```python
df_final.info()
```

```python
df_final.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
df_final.head(1)
```

```python
df_final['Arrival_Time'] = df_final['Arrival_Time'].apply(lambda x:x.split(' ')[0])

#df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final.head(1)
```

```python
df_final['Arrival_Hour'] = df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arrival_Min'] = df_final['Arrival_Time'].str.split(':').str[1]
```

```python
df_final.head(1)
```

```python
df_final['Arrival_Hour'] = df_final['Arrival_Hour'].astype(int)
df_final['Arrival_Min'] = df_final['Arrival_Min'].astype(int)
```

```python
df_final.info()
```

```python
df_final.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
df_final.head(1)
```

```python
df_final['Departure_Hour'] = df_final['Dep_Time'].str.split(':').str[0]
df_final['Departure_Min'] = df_final['Dep_Time'].str.split(':').str[1]

df_final['Departure_Hour'] = df_final['Departure_Hour'].astype(int)
df_final['Departure_Min'] = df_final['Departure_Min'].astype(int)
```

```python
df_final.head(1)
```

```python
df_final.drop('Dep_Time', axis = 1, inplace = True)
```

```python
df_final.head(1)
```

```python
df_final['Total_Stops'].unique()
```

```python
df_final['Total_Stops'] = df_final['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':5, '4 stops':4})
```

```python
df_final['Total_Stops'].unique()
```

```python
df_final['Total_Stops'] = df_final['Total_Stops'].fillna(0)
```

```python
df_final.head(5)
```

```python
df_final['Total_Stops'].unique()
```

```python
df_final['Duration_Hour'] = df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df_final.head(1)
```

```python
df_final.drop(6474,axis=0,inplace = True)
df_final.drop(2660,axis=0,inplace = True)
```

```python
df_final[df_final['Duration_Hour'] == '5m']
```

```python
df_final['Duration_Hour'] = df_final['Duration_Hour'].astype(int)
```

```python
df_final.info()
```

```python
df_final['Duration_Hour']  = df_final['Duration_Hour']*60
```

```python
df_final.head(2)
```

```python
df_final['Duration_Minute'] = df_final['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df_final.head(2)
```

```python
df_final[df_final['Duration_Minute'].isnull()]
```

```python
df_final['Duration_Minute'] = df_final['Duration_Minute'].fillna(0)
```

```python
df_final.drop('Duration', axis=1,inplace=True)
```

```python
#Total Stops is considered, hence we are dropping Route

df_final.drop('Route', axis=1, inplace = True)
```

```python
df_final['Airline'] = labelencoder.fit_transform(df_final['Airline'])
```

```python
df_final['Source'] = labelencoder.fit_transform(df_final['Source'])
df_final['Destination'] = labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info'] = labelencoder.fit_transform(df_final['Additional_Info'])
```
